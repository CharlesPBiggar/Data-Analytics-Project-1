```python
import plotly.express as px
fig = px.histogram(don_df, x="Foundedn Year",
                   y="Funding Total USD",
                   color="coast", histfunc='sum', 
                   title=' Funding by Year / East Coast vs West Coast',
                   orientation='v')

fig.update_layout(xaxis_range=[1990,2015])# trimming the data by years
fig.layout.yaxis.title.text = "Funding Total USD"
fig.show()
```

```python
import plotly.express as px
fig = px.scatter(don_df, x="Category",
                 y="Funding Total USD",
                 color="Category",
                 size='Funding Total USD',
                 hover_data=['name'],
                 title=' Investments by Category')

fig.show()
```
