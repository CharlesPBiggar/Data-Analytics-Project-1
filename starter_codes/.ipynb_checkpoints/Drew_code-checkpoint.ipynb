```python
# read Main Starter File with Funding and weather data combined (CSV)
starter_data = pd.read_csv("../startup_datasets/final_data1.csv")
starter_data.head()
```

```python
#size vs category 
# !!!!main data sheet has to be sorted by funding then category
# !!!!!showing NaN for many values



zed = starter_data.groupby(['city', 'category_list'])['funding_total_usd'].sum()
funds = pd.DataFrame(zed)
f=funds.reset_index()
f.sort_values(by='funding_total_usd', ascending=False)

city_heat = f["city"]
Category = f["category_list"]
z= f['funding_total_usd']

fig = go.Figure(data=go.Heatmap(
        z=z,
        x=city_heat,
        y=Category,
        type='heatmap',
        colorscale='Viridis'))
fig.update_layout(
    font=dict(family="'Oswald', sans-serif", size=5),
     xaxis_nticks=25
)


fig.update_layout(
    title=go.layout.Title(
    text="Total Funding of Category by City",
         font=dict(
                family="'Oswald', sans-serif",
                size=24,
                color="#7f7f7f"
            )
    ),
    xaxis=go.layout.XAxis(
        title=go.layout.xaxis.Title(
            text="City Name",
            font=dict(
                family="'Oswald', sans-serif",
                size=12,
                color="#7f7f7f"
            )
        )
    ),
    yaxis=go.layout.YAxis(
        title=go.layout.yaxis.Title(
            text="Category",
            font=dict(
                family="'Oswald', sans-serif",
                size=12,
                color="#7f7f7f"
            )
        )
    )
)

fig.show()
```

```python
gmaps.configure(api_key=gkey)


mapsdf = starter_data[['city','Lat','Lng']]
mapsdf = pd.DataFrame(mapsdf)
mapsdf.sort_values("city", inplace = True)
mapsdf.drop_duplicates(subset ="city", 
                     keep = 'first', inplace = True)

locations = mapsdf[['Lat','Lng']]
weight = starter_data.groupby('city')['funding_total_usd'].sum()

fig = gmaps.figure()
heatmap_layer = gmaps.heatmap_layer(
    locations, weight,
    max_intensity=1000000, point_radius=1.0
)

heatmap_layer.max_intensity = 1000000000
heatmap_layer.point_radius = 9
fig.add_layer(heatmap_layer)
fig
```

```python
url = "https://raw.githubusercontent.com/plotly/datasets/master/gapminderDataFiveYear.csv"
dataset = pd.read_csv(url)

years = starter_data['founded_at']

# make list of categories = continents
categories = starter_data["category_list"]

        
# make figure
fig_dict = {
    "data": [],
    "layout": {},
    "frames": []
}

# fill in most of layout
fig_dict["layout"]["xaxis"] = {"range": [30, 85], "title": "Year Founded"}
fig_dict["layout"]["yaxis"] = {"title": "Funding Totals", "type": "log"}
fig_dict["layout"]["hovermode"] = "closest"
fig_dict["layout"]["sliders"] = {
    "args": [
        "transition", {
            "duration": 400,
            "easing": "cubic-in-out"
        }
    ],
    "initialValue": "1952",
    "plotlycommand": "animate",
    "values": years,
    "visible": True
}
fig_dict["layout"]["updatemenus"] = [
    {
        "buttons": [
            {
                "args": [None, {"frame": {"duration": 500, "redraw": False},
                                "fromcurrent": True, "transition": {"duration": 300,
                                                                    "easing": "quadratic-in-out"}}],
                "label": "Play",
                "method": "animate"
            },
            {
                "args": [[None], {"frame": {"duration": 0, "redraw": False},
                                  "mode": "immediate",
                                  "transition": {"duration": 0}}],
                "label": "Pause",
                "method": "animate"
            }
        ],
        "direction": "left",
        "pad": {"r": 10, "t": 87},
        "showactive": False,
        "type": "buttons",
        "x": 0.1,
        "xanchor": "right",
        "y": 0,
        "yanchor": "top"
    }
]

sliders_dict = {
    "active": 0,
    "yanchor": "top",
    "xanchor": "left",
    "currentvalue": {
        "font": {"size": 20},
        "prefix": "Year:",
        "visible": True,
        "xanchor": "right"
    },
    "transition": {"duration": 300, "easing": "cubic-in-out"},
    "pad": {"b": 10, "t": 50},
    "len": 0.9,
    "x": 0.1,
    "y": 0,
    "steps": []
}

# make data
year = 1952
for category in categories:
    dataset_by_year = dataset[dataset["year"] == year]
    dataset_by_year_and_cont = dataset_by_year[
        dataset_by_year["category"] == category]

    data_dict = {
        "x": list(dataset_by_year_and_cont["foundyear"]),
        "y": list(dataset_by_year_and_cont["funding"]),
        "mode": "markers",
        "text": list(dataset_by_year_and_cont["category"]),
        "marker": {
            "sizemode": "area",
            "sizeref": 200000,
            "size": list(dataset_by_year_and_cont["pop"])
        },
        "name": category
    }
    fig_dict["data"].append(data_dict)

# make frames
for year in years:
    frame = {"data": [], "name": str(year)}
    for category in categories:
        dataset_by_year = dataset[dataset["foundyear"] == int(year)]
        dataset_by_year_and_cont = dataset_by_year[
            dataset_by_year["category"] == category]

        data_dict = {
            "x": list(dataset_by_year_and_cont["lifeExp"]),
            "y": list(dataset_by_year_and_cont["funding"]),
            "mode": "markers",
            "text": list(dataset_by_year_and_cont["category"]),
            "marker": {
                "sizemode": "area",
                "sizeref": 200000,
                "size": list(dataset_by_year_and_cont["pop"])
            },
            "name": category
        }
        frame["data"].append(data_dict)

    fig_dict["frames"].append(frame)
    slider_step = {"args": [
        [year],
        {"frame": {"duration": 300, "redraw": False},
         "mode": "immediate",
         "transition": {"duration": 300}}
    ],
        "label": year,
        "method": "animate"}
    sliders_dict["steps"].append(slider_step)


fig_dict["layout"]["sliders"] = [sliders_dict]

fig = go.Figure(fig_dict)

fig.show()
```
